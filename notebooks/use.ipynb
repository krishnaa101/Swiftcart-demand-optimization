# Pseudo Logic of how to use model for future Orders.

- Step 1 -- Roughly estimate units to be ordered (say est_order = x units).
- Step 2 -- Provide other input features required for model along with units_ordered.
- Step 3 -- We will get a prediction of units sold (say y units).
- Step 4 -- We set a safety buffer (small value of say 0.05).
- Step 4 -- if x>>y (it means overstocking problem) then we should order y(1+safety buffer).
- Step 5 -- or if abs(x-y) <5(say) then we should order x/y(whichever is bigger).
- Step 6 -- or if  y==x (understocking) then we can order y(1+2*safety buffer) or add something else to y depending on company .

NOTE - Above written logic is just a way to use the model. Company can acquire other optimization techniques also.

# Code for using the model 

```python
import joblib
import pandas as pd

model = joblib.load("inventory_optimisation.pkl")

# Example: predict units sold for a single SKU-store-day
features = pd.DataFrame([{
    "Category_Enc": 1,        # 0=Bakery, 1=Dairy, 2=Pantry, 3=Produce, 4=Snacks
    "Unit_Cost": 45.0,
    "Retail_Price": 120.0,
    "Units_Ordered": 100,
    "DayOfWeek": 3,           # 0=Monday ... 6=Sunday
    "Month": 7,
    "Quarter": 3,
    "WeekOfYear": 28,
    "IsWeekend": 0,
    "IsPerishable": 1,
    "gross_margin": 0.625
}])

predicted_units_sold = model.predict(features)[0]
safety_buffer = 0.05
optimal_order = round(predicted_units_sold * (1 + safety_buffer))

print(f"Forecast: {predicted_units_sold:.1f} units")
print(f"Optimal order: {optimal_order} units")
```
